# Find the top 10 revenue generating products?

```python
# Group by Product and sum the revenue
revenue_by_product = df.groupby('product_id')['sale_price'].sum().reset_index()

# Sort the products by revenue in descending order and get the top 10
top_10_products = revenue_by_product.sort_values(by='sale_price', ascending=False).head(10).reset_index(drop = True)

# Display the top 10 revenue-generating products
print(top_10_products)
```

```python
query = """select distinct product_id, sum(sale_price) revenue
from global_sale_data
group by product_id
order by revenue desc
limit 10;"""

cur.execute(query)

data = cur.fetchall()

df1 = pd.DataFrame(data, columns = ['product_id', 'revenue'])

df1
```

# Find the top 5 highest selling product in each region?

```python
# Group by 'region' and 'product_id', and sum the sale_price for each group
grouped_sales = df.groupby(['region', 'product_id'])['sale_price'].sum().reset_index()

# Sort by 'region' and 'sale' in descending order
grouped_sales = grouped_sales.sort_values(['region', 'sale_price'], ascending=[True, False])

# Get the top 5 products for each region
top_5_products_per_region = grouped_sales.groupby('region').head(5).reset_index(drop = True)

print(top_5_products_per_region)
```

```python
query = """with sale_rank as (select region, product_id, sum(sale_price) as sale,
rank() over(partition by region order by sum(sale_price) desc) as rn
from global_sale_data
group by region, product_id)

select region, product_id, sale, rn
from sale_rank
where rn <= 5
order by region, rn;"""

cur.execute(query)

data = cur.fetchall()

df2 = pd.DataFrame(data, columns = ['region', 'product_id', 'sale', 'rn'])

df2
```

# Find month over month growth comparison for 2022 vs 2023 ?

```python
query = """with cte as (select year(order_date) as years, month(order_date)as months, sum(sale_price) as sale
from global_sale_data
group by years, months)

select months,
sum(case when years =2022 then sale else 0 end) as sale_2022,
sum(case when years =2023 then sale else 0 end) as sale_2023
from cte
group by months
order by months;"""

cur.execute(query)

data = cur.fetchall()

df3 = pd.DataFrame(data, columns = ['months', 'sale_2022', 'sale_2023'])

df3
```

# For each category which month has highest sale?

```python
query = """WITH MonthlySales AS (
    SELECT 
        category,
        EXTRACT(MONTH FROM order_date) AS month,
        SUM(sale_price) AS total_sales
    FROM 
        global_sale_data
    GROUP BY 
        category, month
),
RankedSales AS (
    SELECT 
        category,
        month,
        total_sales,
        ROW_NUMBER() OVER (PARTITION BY category ORDER BY total_sales DESC) AS sales_rank
    FROM 
        MonthlySales
)
SELECT 
    category, 
    month, 
    total_sales
FROM 
    RankedSales
WHERE 
    sales_rank = 1
ORDER BY 
    category;"""

cur.execute(query)

data = cur.fetchall()

df4 = pd.DataFrame(data, columns = ['category', 'months', 'total_sales'])

df4
```

# Which sub category had highest growth by profit in 2023 compare to 2022?

```python
query = """WITH cte AS (
    SELECT 
        sub_category, 
        YEAR(order_date) AS order_year,
        SUM(sale_price) AS sales
    FROM 
        global_sale_data
    GROUP BY 
        sub_category, YEAR(order_date)
),
cte2 AS (
    SELECT 
        sub_category,
        SUM(CASE WHEN order_year = 2022 THEN sales ELSE 0 END) AS sales_2022,
        SUM(CASE WHEN order_year = 2023 THEN sales ELSE 0 END) AS sales_2023
    FROM 
        cte
    GROUP BY 
        sub_category
)
SELECT 
    *,
    (sales_2023 - sales_2022) AS sales_difference
FROM 
    cte2
ORDER BY 
    sales_difference DESC;
"""

cur.execute(query)

data = cur.fetchall()

df5 = pd.DataFrame(data, columns = ['sub_category', 'sale_2022', 'sale_2022', 'sales_difference'])

df5
```
